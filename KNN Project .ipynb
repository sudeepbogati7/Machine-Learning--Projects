# K Nearest Neighbor Project 


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
import seaborn as sns
import sklearn
```

```python
# import and overview of the csv file
```

```python
data = pd.read_csv('KNN_Project_Data')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.pairplot(data)
```

# Standardize the variables 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data.drop('TARGET CLASS', axis=1))
```

#### transform the features to a scaled version

```python
scaled_feature = scaler.transform(data.drop('TARGET CLASS',axis=1))
```

### Converting the scaled_features to a dataframe

```python
feat_data = pd.DataFrame(scaled_feature, columns = data.columns[:-1] )
feat_data.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = feat_data
y = data['TARGET CLASS']

X_train ,X_test, y_train , y_test = train_test_split(X,y , test_size=0.3, random_state= 101)
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("confusion metrics : \n", confusion_matrix(y_test , pred),'\n')
print("Classification report : \n", classification_report(y_test,pred))
```

### Choosing a k value ...............

```python
error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(16,10))
plt.title("Error rate vs K values")
plt.xlabel('K values')
plt.ylabel('error rate ')
plt.grid(True)
plt.plot(range(1,50), error_rate, 'b--', marker= 'o', markerfacecolor = 'r', markersize=5,markeredgecolor = 'r')
```

#### From above graph , already the error rate is quite high and at k value 31/32 the error rate is comparatively low 

```python
# Hence using the k-value = 32

knn = KNeighborsClassifier(n_neighbors=32)
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```
